# Combinaison de transformateurs et d'estimateurs dans un pipeline

**StandardScaler (Standarisation) :**
Le StandardScaler est une technique de prétraitement des données qui vise à centrer et à réduire les caractéristiques d'un jeu de données. Cela signifie que chaque caractéristique aura une moyenne de zéro et un écart-type de un. La formule utilisée pour le StandardScaler est la suivante :
$z = (x_i - μ) / σ$

z : la valeur transformée <br>
x_i : la valeur d'origine <br>
μ : la moyenne de la caractéristique<br>
σ : l'écart-type de la caractéristique<br>

En soustrayant la moyenne de la valeur d'origine et en divisant par l'écart-type, nous obtenons la valeur transformée, qui est centrée et réduite. Cela permet d'harmoniser les échelles des différentes caractéristiques et facilite ainsi leur comparaison et leur utilisation dans les modèles d'apprentissage automatique.

**PCA (Analyse en composantes principales) :**
Le PCA est une technique d'analyse de données qui vise à réduire la dimensionnalité d'un jeu de données tout en conservant autant que possible l'information qu'il contient. Il permet de transformer un ensemble de variables corrélées en un nouvel ensemble de variables non corrélées, appelées composantes principales. La formule pour le calcul du PCA est la suivante :
$PC = X * V$

PC : la matrice des composantes principales
X : la matrice des données originales
V : la matrice des vecteurs propres (ou vecteurs de chargement)
Pour effectuer le PCA, nous multiplions la matrice des données originales (X) par la matrice des vecteurs propres (V). Les composantes principales résultantes sont les colonnes de la matrice transformée (PC).<br>

Les composantes principales sont triées par ordre décroissant de leur variance, ce qui signifie que les premières composantes principales capturent la plus grande partie de la variance totale des données.<br>

En réduisant la dimensionnalité, le PCA peut faciliter la visualisation des données et améliorer les performances des modèles d'apprentissage automatique en éliminant les redondances et le bruit dans les données.

````python

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
pipe_lr = make_pipeline(

    StandardScaler(),
    PCA(n_components=2),
    LogisticRegression()
    )



pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
test_acc = pipe_lr.score(X_test, y_test)
print(f'Test accuracy: {test_acc:.3f}')
````
