# K-Means Cluster Analysis

**Les algorithmes d'apprentissage non supervisé** tentent d'apprendre des modèles à partir d'ensembles de données non étiquetés, en découvrant des similarités ou des régularités. Les tâches courantes de l'apprentissage non supervisé comprennent le regroupement (clustering) et l'association. Les algorithmes de regroupement, comme K-means, cherchent à découvrir des similarités au sein de l'ensemble de données en regroupant les objets de telle manière que les objets d'un même cluster soient plus similaires entre eux que par rapport aux objets d'un autre cluster. Le regroupement en clusters se fait en utilisant des critères tels que les plus petites distances, la densité des points de données, les graphiques ou diverses distributions statistiques.

K-means regroupe des points de données similaires ensemble en minimisant la distance moyenne entre les points géométriques. Pour ce faire, il partitionne de manière itérative les ensembles de données en un nombre fixe (le K) de sous-groupes (ou clusters) non chevauchants, où chaque point de données appartient au cluster dont le centre de cluster moyen est le plus proche.

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

La fonction **make_blobs** est utilisée pour générer des données simulées en forme de **"blobs"** dans le but de tester des algorithmes d'apprentissage automatique.

Lorsque vous utilisez la fonction make_blobs, vous pouvez spécifier plusieurs paramètres pour contrôler la forme et les propriétés des données générées, tels que le nombre de clusters, le nombre de points de données, les centres des clusters, la dispersion, etc.


```python
from sklearn.datasets import make_blobs # pour creer des donnees dans le cadre de simuler le fonctionnement de K moyenne.
```

Pour creer des donnees on utilise la fonction make_bolbs on passe les parametres suivants:


*   n_sample: le numbre d'element que vous souhaitez avoir
*   n_features: Nombre de colonne que vous souhaitez
* centers: le nombre de repartition des points(group a former)
* Cluser_std: l'ecart-type de donnee.



```python
data = make_blobs(n_samples=300,n_features=2,centers=4,cluster_std=1.8,
                 random_state=101)
```

Pour faire la representation graphique de niage des points on utilise plt.scatter

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

En résumé, cette ligne de code crée un graphique de dispersion avec les données de data[0] en utilisant les valeurs de la première colonne pour l'axe X, les valeurs de la deuxième colonne pour l'axe Y, et les couleurs des points sont déterminées par les valeurs de data[1]. Cela permet de visualiser la répartition des points et leur association avec différentes classes ou étiquettes.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
fig ,(ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6),)

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('K Means')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap='rainbow', marker='^')  # Use marker='o' for circles
ax2.set_title("Original")
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow', marker='s')  # Use marker='s' for squares
```

## Importance de K-means
Le clustering K-means est une technique de regroupement de données largement utilisée en apprentissage automatique et en exploration de données.


1. **Segmentation des clients** : Le clustering K-means peut être utilisé pour segmenter les clients en groupes homogènes en fonction de leurs caractéristiques ou comportements communs. Par exemple, dans le domaine du marketing, cela peut aider à identifier différents segments de clients pour cibler des campagnes publicitaires spécifiques à chaque groupe. Cette segmentation peut permettre d'adapter les stratégies de marketing en fonction des besoins et des préférences de chaque segment.

2. **Compression des données** : Le clustering K-means peut être utilisé pour compresser des ensembles de données volumineux. En identifiant les groupes similaires dans les données, on peut représenter ces groupes par leurs centroides ou leurs centres, réduisant ainsi la dimension des données.

3. **Analyse de données non étiquetées** : Le clustering K-means est souvent utilisé pour analyser des ensembles de données non étiquetées, c'est-à-dire des données qui ne sont pas préalablement classées ou catégorisées. Il peut être utilisé pour découvrir des structures sous-jacentes, des tendances ou des motifs dans les données. Cette méthode peut être appliquée dans divers domaines, tels que l'analyse des réseaux sociaux, la bioinformatique, l'exploration de données géospatiales, etc.
