# Prédiction d' une variables continues avec l'analyse de régression:

**Introduction à la régression linéaire:**
L'objectif de la régression linéaire est de modéliser la relation entre une ou plusieurs variables explicatives et une variable cible continue.<br>

Contrairement à la classification - une autre sous-catégorie de l'apprentissage supervisé - l'analyse de régression vise à prédire des sorties sur une échelle continue plutôt que des étiquettes de classe catégoriques.

## Régression linéaire simple
L'objectif de la régression linéaire simple (univariée) est de modéliser la relation entre une seule variable explicative (variable x) et une variable cible à valeurs continues (variable y). L'équation d'un modèle linéaire avec une seule variable explicative est définie comme suit : $y = wx + b$

Ici, le paramètre (unité de biais), $b$, représente l'intercepteur de l'axe y, et $w$ est le coefficient de pondération de la variable explicative.

# Régression linéaire multiple

La régression linéaire multiple est une extension de la régression linéaire simple qui permet de modéliser la relation entre plusieurs variables explicatives (également appelées variables indépendantes) et une variable cible continue.

L'equation: $ y = b_0 + b_1*x_1 + b_2*x_2 + ... + b_n*x_n$ <br>

Ou<br>

$y = $$\sum_{i=1}^n W_i X_i + b_0$

**Les hypothèses de la régression linéaire** sont des conditions qui doivent être satisfaites pour obtenir des estimations fiables et interprétables avec ce modèle. Voici les principales hypothèses de la régression linéaire :

1. Linéarité : L'hypothèse de linéarité suppose que la relation entre les variables explicatives et la variable cible est linéaire. Cela signifie que les coefficients de régression demeurent constants et la somme des termes linéaires est utilisée pour modéliser la relation.

2. Homoscédasticité : Cette hypothèse suppose que la variance des résidus (différence entre les valeurs prédites et les valeurs réelles) est constante pour toutes les valeurs des variables explicatives. En d'autres termes, la dispersion des résidus reste la même dans tout l'éventail des valeurs prédites.

3. Indépendance des résidus : Les résidus de la régression doivent être indépendants les uns des autres, ce qui signifie qu'il ne doit pas y avoir de corrélation systématique entre les résidus. Cela garantit que chaque observation apporte une information unique et n'est pas influencée par les autres.

4. Absence de colinéarité : L'hypothèse d'absence de colinéarité suppose que les variables explicatives ne sont pas fortement corrélées entre elles. Une forte corrélation entre les variables explicatives peut entraîner une instabilité dans les coefficients de régression et rendre difficile l'interprétation des effets individuels des variables.

5. Normalité des résidus : Cette hypothèse suppose que les résidus suivent une distribution normale. Cela permet d'obtenir des intervalles de confiance et des tests d'hypothèse valides.

6. Absence d'effets significatifs des valeurs aberrantes : Les valeurs aberrantes sont des observations qui diffèrent considérablement des autres observations. L'hypothèse d'absence d'effets significatifs des valeurs aberrantes suppose que ces observations n'ont pas d'influence disproportionnée sur les résultats de la régression.

Il est important de noter que toutes ces hypothèses doivent être évaluées et vérifiées avant d'interpréter les résultats de la régression linéaire. Dans certains cas, des violations mineures de ces hypothèses peuvent être tolérées, mais des violations majeures peuvent entraîner des estimations biaisées ou peu fiables.

# L'évaluation des performances des modèles de régression linéaire:
Elle permet de mesurer à quel point ces modèles sont capables de prédire de manière précise les valeurs de la variable cible. <br>

Voici quelques métriques couramment utilisées pour évaluer la performance des modèles de régression linéaire :


L'évaluation des performances des modèles de régression linéaire permet de mesurer à quel point ces modèles sont capables de prédire de manière précise les valeurs de la variable cible. Voici quelques métriques couramment utilisées pour évaluer la performance des modèles de régression linéaire :

1. Erreur quadratique moyenne (RMSE) : Le RMSE mesure l'écart moyen entre les valeurs prédites par le modèle et les valeurs réelles de la variable cible. Une valeur de RMSE plus faible indique une meilleure performance du modèle.


2. Erreur absolue moyenne (MAE) : Le MAE calcule la moyenne des écarts absolus entre les valeurs prédites et les valeurs réelles. Il fournit une mesure de l'erreur moyenne du modèle.

3. Coefficient de détermination (R²) : Le coefficient de détermination mesure la proportion de la variance de la variable cible qui peut être expliquée par le modèle. Il indique à quel point les variables explicatives du modèle capturent la variation de la variable cible. Une valeur de R² proche de 1 indique un bon ajustement du modèle.

Il est important de noter que le choix des métriques d'évaluation dépend du contexte et des objectifs spécifiques de la modélisation. Il est recommandé d'utiliser plusieurs métriques pour avoir une vision globale de la performance du modèle.
